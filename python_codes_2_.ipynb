# Question 1

Create a function that takes a string and returns a list of word tokens, ignoring any non-alphanumeric characters. You should use `re.findall` in your function. Test your function with the string "Hello there, how are you? Want to go skiing?" and store the tokens in a variable `tokens`.

```python
import re
def convertString_Tokens(text):
  token1= re.findall(r'\w+', text)
  return token1

text="Hello there, how are you? Want to go skiing?"
tokens=convertString_Tokens(text)
print(tokens)
```

# Stop Words and Lemmas

The following code downloads stop words and lemmas and stores them in the variables `stopwords` and `lemmas`.

```python
import urllib.request, json
with urllib.request.urlopen("https://storage.googleapis.com/wd13/stopwords%20and%20lemmas.json") as url:
  data = json.load(url)
  stopwords = data['stopwords']
  lemmas = data['lemmas']
```

`stopwords` is a list of stop words.

```python
stopwords
```

`lemmas` is a dictionary where the "key" is a word and the "value" is the word's lemma.

```python
lemmas
```

```python
# the lemma of "skiing" is "ski"
lemmas['skiing']
```

# Question 3

Create a function that takes a list of tokens as a parameter and removes any tokens that are in `stopwords`. Test it with your `tokens` variable.

```python
# removing stop words from a list of tokens
def remove_stop(tokens):
  flitered_string = []
  for t in tokens:
    if t not in stopwords:
      flitered_string.append(t)
  print(tokens)
  print(flitered_string)

remove_stop(tokens)
```
